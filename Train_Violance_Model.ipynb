```python
import numpy as np
import pandas as pd

#load data that saved from video

data1 = pd.read_csv("voilence.csv")
data2 = pd.read_csv("walk.csv")
# create label
data1["51"]=1
data2["51"]=0
# total data 
data3 = pd.concat([data1, data2], axis=0)
data3 = data3.set_index(data3.columns[0])

data3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

x=data3.iloc[:,:-1]
y=data3.iloc[:,-1]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30,shuffle=True)
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
import pickle

pickle.dump(model,open("voilance-model.h5","wb"))
```
